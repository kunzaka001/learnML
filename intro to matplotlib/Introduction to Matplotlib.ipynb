```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y);
```

```python
plt.plot();
```

```python
# lst method
fig = plt.figure() # create a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd method (recommanded)
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]); # add some data
```
